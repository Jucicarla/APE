```
import textdistance
import editdistance
```

```
import pandas as pd
```

```
def distance_string(str1, str2):

  editdist = editdistance.eval(str1, str2)
  print("Edit distance: ", editdist)

  jaro = textdistance.jaro(str1, str2)
  print("Jairo: ", jaro)

  haming = textdistance.hamming(str1, str2)
  print("Haming: ", haming)

  hamingDist = textdistance.hamming.distance(str1, str2)
  print( "HamingDist: ", hamingDist)

  hamingSim = textdistance.hamming.similarity(str1, str2)
  print( "HamingSim: ", hamingSim)

  hamingNDist = textdistance.hamming.normalized_distance(str1, str2)
  print( "HamingNDist", hamingNDist)

  hamingNSim = textdistance.hamming.normalized_similarity(str1, str2)
  print( "HamingNSim: ", hamingNSim)

  hamingDistq = textdistance.Hamming(qval=2).distance (str1, str2)
  print( "HamingDist: ", hamingDistq)
 
```

```
distance_string('etxt', 'text')
```

```
def gramas_string(str1, str2):
  jaccard = textdistance.jaccard(str1, str2)
  print('jaccard', jaccard)

  sorensen = textdistance.sorensen_dice(str1, str2)
  print("sorensen", sorensen)
```

```
tokens_1 = 'hello wordl'#.split()
tokens_2 = 'world new hello'

gramas_string(tokens_1, tokens_2)
```

```
tokens_1 = 'hello wordl'.split()
tokens_2 = 'world new hello'.split()

gramas_string(tokens_1, tokens_2)
```
